```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('historic_groceries.csv')
df
```

```python
df.info()
```

```python
df = df.drop('Sunflower-oil',axis=1)
df
```

```python
df['month'] = df.Month.str.split('-').map(lambda x: x[0])
df['year'] = df.Month.str.split('-').map(lambda x: x[1])
df  = df.drop('Month',axis=1)
df
```

```python
df = df.drop([0,1,2,351,352,353,354,355,356,357,358,359],axis=0)
df
```

```python
sales_by_year = df.groupby('year').sum()
sales_by_year
```

```python
products = sales_by_year.columns
for product in products:
    plt.figure(figsize=(20,10))
    sns.barplot(x=sales_by_year.index,y=product,data=sales_by_year)
    plt.title(product)
```

```python
avg_by_month = df.groupby('month').mean()
avg_by_month
```

```python
items = avg_by_month.columns
for item in items:
    plt.figure(figsize=(20,10))
    sns.barplot(x=avg_by_month.index,y=item,data=avg_by_month)
    plt.title(item)
```
